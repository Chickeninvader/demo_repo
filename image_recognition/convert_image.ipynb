```python
import glob
import cv2
import numpy as np
from keras.preprocessing.image import array_to_img
from math import sqrt

train_cat = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/train/cat/*.jpg"
train_dog = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/train/dog/*.jpg"
test_cat = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/test/cat/*.jpg"
test_dog = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/test/dog/*.jpg"
dev_cat = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/dev/cat/*.jpg"
dev_dog = "D:/anhbin/programming_python/demo_repo/deep_learning/dogvscat/dev/dog/*.jpg"

def convert_image(directory):
    """
    Argument: 
    directory -- list of directory where we want to convert picture (in this program, we limit the use to
    only 2 file contains training and test set of cats and dog)

    Return:
    image1 -- 2D array contain detail of pictures (size_of_picture*3, number_of_examples)
    outcome -- an array contain information of picture (1, number_of_examples)
    We choose 0 = cat, 1 = dog
    """
    image1 = []
    outcome = []
    for dir in directory:
        for file in glob.glob(dir):
            image = cv2.imread(file)
            # We resize every picture to 100*100 and divide all elements so that all of them is less than 1
            try:
                image = cv2.resize(image, (75, 75),
                                   interpolation=cv2.INTER_AREA)/255.
            except:
                break
            image1.append(image)
            if dir == train_cat or dir == test_cat or dir == dev_cat:
                outcome.append(0)
            else:
                outcome.append(1)

    outcome = np.array(outcome)
    outcome = outcome.reshape(1, outcome.shape[0])
    image1 = np.array(image1)
    image1 = np.reshape(image1, (image1.shape[0], -1))
    image1 = np.transpose(image1)
    print(image1.shape)
    print(outcome.shape)

    return image1, outcome


def show_image(index, array_of_image):
    i = int(sqrt(array_of_image.shape[0]/3))
    image = np.reshape(array_of_image[:, index], (i, i, 3))*255
    image = array_to_img(image)
    image.show()

```
